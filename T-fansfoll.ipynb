```python
from tqdm import tqdm 
import time
import pandas as pd
import numpy as np
import requests, json, csv, time, os,sys,os,math,datetime
from math import ceil
from jsonsearch import JsonSearch
```

```python
#aa = pd.read_csv('/Users/tangyilin/Desktop/user.csv')
```

```python
#粉丝ID
t1= time.time()#开始时间
list1 = ['9650112278165562']
#print(list1[0])
a=0
while a>=0:
    url = "https://i.eastmoney.com/api/guba/userfans?uid="+list1[a]#粉丝
    #url = "https://i.eastmoney.com/api/guba/userFollowPerson?uid="+df2['user_id'][a]
     # 发起请求
    response = requests.get(url)
    response.encoding='utf-8'
    # 获取响应,提取粉丝列表
    js_data = response.json()
    if response.json()["re"] == False:
        print ("sssssssssssssssssssssss")
        time.sleep(1)
        continue   
    else: js_data = response.json()["result"]
    #aa_list = js_data['result']
    n = ceil(js_data['count']/10)+1#list1为int
#         print(list1) 
    for i in tqdm(range(1,n)):
        url2 = url+"&pageindex="+str(i)
    #         pint(url2)
        response = requests.get(url2)
        response.encoding='utf-8'
        js_data = response.json()
        jsondata = JsonSearch(object=js_data, mode='j')#避免回圈直接取得json对应位置的值
        jj = jsondata.search_all_value(key='user_id')
    if len(jj)==0:
            a += 1
    if len(jj)!=0:
        list1=list1+jj
        df = pd.DataFrame(jj)
        df.columns = ['user_id'] + df.columns[1:].tolist()
        df3 = df[~df['user_id'].isin(df['user_id'])]
    if len(df3['user_id'])==0:
        a += 1
    #if len(df3['user_id'])!=0: 
    #   ctz_index.insert_DataFrame(db,collection,df3)

    a+=1
    if len(list1)>=200:
        break  #控制数量
    #     stop
t2= time.time()
print (t2-t1) #进度条
```
